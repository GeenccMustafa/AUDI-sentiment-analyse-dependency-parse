```python
import re

def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
# Initialize the spaCy engine for English language
nlp_eng = spacy.load("en_core_web_sm")
# Initialize the spaCy engine for English language
nlp_eng = spacy.load("en_core_web_sm")
def lemma_eng(review):
    lemma_sent_eng = []
    review = ' '.join(review.split())
    doc = nlp_eng(review)
    for sent in doc.sents:
        new_doc = nlp_eng(sent.text)
        new_sent = " "
        new_sent = " ".join([token.lemma_ for token in new_doc])
        lemma_sent_eng.append(new_sent)
    adjusted_review = " ".join(lemma_sent_eng)
    return adjusted_review
```

```python
import nltk
nltk.download('punkt')
def stop_eng(review):
    eng_stopwords = nlp_eng.Defaults.stop_words
    text_tokens = word_tokenize(review)
    tokens_without_sw = [word for word in text_tokens if not word in eng_stopwords]
    adjusted_review = " ".join(word for word in tokens_without_sw)
    return adjusted_review
```

```python
def print_emotion_results(emo_results):
    emo_results = emo_results[0]
    printed_str = ''
    highest_emo = max(emo_results, key=lambda x:x['score'])['label']
    for emo in emo_results:
        if emo['label'] == highest_emo:
            printed_str += colored(f"{emo['label']}: {float(emo['score']):.3f}  ", 'red')
        elif emo['score'] >= 0.2:
            printed_str += colored(f"{emo['label']}: {float(emo['score']):.3f}  ", 'blue')
        else:
            printed_str += f"{emo['label']}: {float(emo['score']):.3f}  "

    return printed_str
```

```python
def get_emotion_score_list(emo_results):
    score_list = []
    emo_results = emo_results[0]
    for emo in emo_results:
        score_list.append(emo['score'])
    return score_list
```

```python
import collections
def get_emotion_score_list4(emo_results):
  
    emotions_dict = {}
    emo_results = emo_results[0]

    for emo in emo_results:
        if emo["label"] == "joy":
            emotions_dict["joy"] = emo['score']
        elif emo["label"] == "anger":
            emotions_dict["anger"] = emo['score']
        elif emo["label"] == "optimism":
            emotions_dict["optimism"] = emo['score']
        else:
            emotions_dict["sadness"] = emo['score']
    emotions = collections.OrderedDict(sorted(emotions_dict.items()))
    return list(emotions.values())
```

```python
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
model.save_pretrained(MODEL)

def get_emotion_model4(text):
    result = [[]] 
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    color = True
    for i in range(scores.shape[0]):
        l = labels[ranking[i]]
        s = scores[ranking[i]]
        result[0].append({"label":l, "score":s})
    return result

 # colored(f"{emo['label']}: {float(emo['score']):.3f}  ", 'blue')
```
